```python

for i in haberdf['positive_nodes']:
    if i in range(0,11):
        haberdf['positive_nodes'].loc[haberdf['positive_nodes']==i]=10
    elif i in range(11,21):
        haberdf['positive_nodes'].loc[haberdf['positive_nodes']==i]=20
    elif i in range(21,31):
         haberdf['positive_nodes'].loc[haberdf['positive_nodes']==i]=30
    elif i in range(31,41):
         haberdf['positive_nodes'].loc[haberdf['positive_nodes']==i]=40
    else:
        haberdf['positive_nodes'].loc[haberdf['positive_nodes']==i]=50
```

```python
coun,bine=np.histogram(yesdf['positive_nodes'],bins=5,density=True)
pdf=coun/sum(coun)
cdf=np.cumsum(pdf)
plt.plot(bine[1:],pdf)
plt.plot(bine[1:],cdf)
```

```python
#90% of survived pateints have 17 positive nodes
```

```python
coun,bine=np.histogram(nodf['positive_nodes'],bins=5,density=True)
pdf=coun/sum(coun)
cdf=np.cumsum(pdf)
plt.plot(bine[1:],pdf)
plt.plot(bine[1:],cdf)
plt.line
```

```python
#70% of unsurvived pateints have 20 positive nodes
#90% of unsurvived pateints have positive nodes<22
```
