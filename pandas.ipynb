#Merge now

```python
temp_df=pd.DataFrame({
    'city':['mum','del','bang'],
    'temp':[21,43,67]    
})
hum_df=pd.DataFrame({
    'city':['mum','del','bang','fas'],
    'hum':[65,32,12,12]
})
```

```python
 #we join/merge them
```

```python
merge_df=pd.merge(temp_df,hum_df,on='city')
merge_df
#fas of hum_df is not merged
```

```python
#to get fas is
```

```python
merge_out_df=pd.merge(temp_df,hum_df,on='city',how='outer')
merge_out_df
```
